```python
url_1 = "https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/QXrVDUoQvFhIcg"
olist_customers_dataset = pd.read_csv(url_1)


url_2 = "https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/0zWmh1bYP6REbw"
olist_orders_dataset = pd.read_csv(url_2)

url_3 = "https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/xCsQ0FiC-Ue4Lg"
olist_order_items_dataset = pd.read_csv(url_3)

```

```python
olist_orders_dataset['order_purchase_timestamp']= pd.to_datetime(olist_orders_dataset.order_purchase_timestamp)\
                                                .dt.strftime('%Y-%m-%d')
```

```python
merge_olist_orders_dataset =  olist_orders_dataset.merge(olist_customers_dataset, on='customer_id')
merge_olist_orders_dataset = merge_olist_orders_dataset[{'customer_unique_id', 'order_id',
                                                         'order_status','order_purchase_timestamp'}]
year_month_data = merge_olist_orders_dataset.query("order_status != ('created','unavailable','canceled') ")
```

```python
year_month_date_ = year_month_data.sort_values("order_purchase_timestamp")
first_purchases = year_month_date_.groupby('customer_unique_id',as_index=False)\
                                .first() #получили датафрейм с датами первых покупок клиентов. 
first_purchases['order_purchase_timestamp'] =  pd.to_datetime(first_purchases['order_purchase_timestamp'])
first_purchases = first_purchases.sort_values("order_purchase_timestamp")
first_purchases['year'] =  pd.to_datetime(first_purchases['order_purchase_timestamp'])\
                                              .dt.strftime('%Y')
first_purchases['month'] =  pd.to_datetime(first_purchases['order_purchase_timestamp'])\
                                              .dt.strftime('%m')
first_purchases['name_month'] =  pd.to_datetime(first_purchases['order_purchase_timestamp'])\
                                              .dt.month_name()
first_purchases_df = first_purchases.query("year == '2017'") #датафрейм с отфильтрованными покупками с нужным годом. 
```

```python
#есть дф с названием когорты и первыми продажами. Месяц покупки и есть название когорты. Есть дф, со всеми покупками. 
#Когда мы вычитаем дату покупки из даты первой покупки, мы получаем месяц, через какое время было совершена повторная покупка. 
#количество покупок в определенный месяц спустя какое то количество месяцев - это сгруппированный по ид , затем по месяцу и
#посчитанное количество ордеров
```

```python
main_cohorts = first_purchases_df[{'name_month','customer_unique_id'}]\
                    .groupby('name_month')
cohorts_year_month_data = year_month_data[{'customer_unique_id','order_purchase_timestamp'}]
cohorts_year_month_data['order_purchase_timestamp'] =  pd.to_datetime(cohorts_year_month_data['order_purchase_timestamp'])
cohorts_year_month_data = cohorts_year_month_data.sort_values("order_purchase_timestamp")
cohorts_year_month_data['year'] =  pd.to_datetime(cohorts_year_month_data['order_purchase_timestamp'])\
                                              .dt.strftime('%Y')
cohorts_year_month_data['month'] =  pd.to_datetime(cohorts_year_month_data['order_purchase_timestamp'])\
                                              .dt.strftime('%m')
cohorts_year_month_data['name_month'] =  pd.to_datetime(cohorts_year_month_data['order_purchase_timestamp'])\
                                              .dt.month_name()
# датафрейм со всеми продажами со всеми годами
```

```python
#Делаем мердж датасета со всеми покупками с датасетом с нашими когортами
aaa = cohorts_year_month_data.merge(first_purchases_df, on='customer_unique_id',how='inner')
#Отсекаем года меньше 2017 
bbb = aaa.query("order_purchase_timestamp_y >='2017-01-01' & order_purchase_timestamp_y >='2017-01-01'")
bbb[['year_x','month_x','year_y','month_y']] = bbb[['year_x','month_x','year_y','month_y']].astype(int)
```

```python
#В новом датасете, если год 2018 и старше - прибавляем 12 месяцев к стобцу месяца, чтобы не получилось отрицательное число при вычитании
for index, row in bbb.iterrows():
    if int(row['year_x'])>2017:
        bbb.at[index, 'month_x'] += 12

```

```python
#Вычитаем из месяцев покупок левой части мерджа правую часть мерджа (там у нас месяцы первой покупки пользователя) 
bbb['difference'] = (bbb['month_x'] - bbb['month_y'])
hz = bbb.groupby(['difference','name_month_x'],as_index=False).agg({'order_id':'count'})
hz_2 = hz.pivot(columns='difference', index='name_month_x', values='order_id').reset_index()
hz_2
```

```python
#ставим месяцы по порядку календаря
reind = [4,3,7,0,8,6,5,1,11,10,9,2]
hz_2 = hz_2.reindex(reind)
hz_2.reset_index(inplace=True)
hz_2.drop(columns = ['index'], axis=1, inplace=True)
```
